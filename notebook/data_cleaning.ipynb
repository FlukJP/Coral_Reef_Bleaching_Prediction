```python
from ydata_profiling import ProfileReport
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import  keras
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df_ROC = pd.read_csv("../data/realistic_ocean_climate_dataset.csv")
df_GBE = pd.read_csv("../data/global_bleaching_environmental.csv")
```

```python
profile_ROC = ProfileReport(df_ROC, title="Realistic Ocean Climate Dataset Profiling Report", html={"style": {"full_width": True}}, sort=None)
profile_GBE = ProfileReport(df_GBE, title="Global Bleaching Events Dataset Profiling Report", html={"style": {"full_width": True}}, sort=None)
```

```python
columns_drop_ROC = [ 'Date', 'Location', 'Species Observed']
df_clean_ROC = df_ROC.drop(columns=columns_drop_ROC, axis=1)
```

```python
df_clean_ROC['Bleaching Severity'] = df_clean_ROC['Bleaching Severity'].fillna(df_clean_ROC['Bleaching Severity'].mode()[0])
```

```python
df_clean_ROC['Bleaching Severity'] = (df_clean_ROC['Bleaching Severity'].astype(str).str.strip().str.lower().map({'low': 0, 'medium': 1, 'high': 2}))
```

```python
core_features = [ 'Depth_m', 'Distance_to_Shore', 'Turbidity', 'Windspeed', 'Cyclone_Frequency', 'Temperature_Maximum', 'SSTA_DHW', 'Percent_Bleaching' ]
df_clean_GBE = df_GBE[core_features].copy()
```

```python
df_clean_GBE = df_clean_GBE.replace('nd', np.nan)
```

```python
numeric_columns = [
    'Depth_m', 'Distance_to_Shore', 'Turbidity', 'Windspeed', 
    'Cyclone_Frequency', 'Temperature_Maximum', 'SSTA_DHW', 'Percent_Bleaching'
]
```

```python
for col in numeric_columns:
    df_clean_GBE[col] = pd.to_numeric(df_clean_GBE[col], errors='coerce')
```

```python
df_clean_GBE.isnull().any()
```

```python
df_clean_GBE.nunique()
```

```python
profile_clean_GBE = ProfileReport(df_clean_GBE, title="Global Bleaching Events Dataset Profiling Report", html={"style": {"full_width": True}}, sort=None)
```

```python
df_clean_GBE.describe(include = 'all')
```

```python
profile_clean_GBE.to_notebook_iframe()
```

```python

mean_cols = ['Windspeed', 'Temperature_Maximum']
for col in mean_cols:
    df_clean_GBE[col] = df_clean_GBE[col].fillna(df_clean_GBE[col].mean())
median_cols = ['Depth_m', 'Distance_to_Shore', 'Turbidity', 'SSTA_DHW']
for col in median_cols:
    df_clean_GBE[col] = df_clean_GBE[col].fillna(df_clean_GBE[col].median())
```
